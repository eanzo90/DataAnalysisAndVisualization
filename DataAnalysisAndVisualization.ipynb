### 1. Setup & Load Data

#### Note: Creating a dummy `system_logins.csv` file.

Since the `system_logins.csv` file was not provided, I will create a dummy one for demonstration purposes. **If you have your own `system_logins.csv` file, please upload it to your Colab environment and remove or comment out the following code.**

```python
# Import necessary libraries for Setup & Load Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import random

# Create a dummy system_logins.csv file for demonstration
num_records = 1000

timestamps = [datetime.now() - timedelta(minutes=random.randint(0, 10000)) for _ in range(num_records)]
user_ids = [f'user_{random.randint(1, 50)}' for _ in range(num_records)]
ip_addresses = [f'{random.randint(1, 255)}.{random.randint(1, 255)}.{random.randint(1, 255)}.{random.randint(1, 255)}' for _ in range(num_records)]
countries = random.choices(['USA', 'Canada', 'UK', 'Germany', 'China', 'India'], k=num_records)
success = random.choices([True, False], weights=[0.85, 0.15], k=num_records)
os = random.choices(['Windows', 'macOS', 'Linux', 'Android', 'iOS'], k=num_records)

dummy_df = pd.DataFrame({
    'timestamp': timestamps,
    'user_id': user_ids,
    'ip_address': ip_addresses,
    'country': countries,
    'success': success,
    'os': os
})
dummy_df.to_csv('system_logins.csv', index=False)

print("Dummy 'system_logins.csv' created.")

# Load system_logins.csv into a DataFrame
try:
    df = pd.read_csv('system_logins.csv')
    print("Successfully loaded 'system_logins.csv'")
except FileNotFoundError:
    print("Error: 'system_logins.csv' not found. Please ensure the file is uploaded or created.")
    # If the file is not found, create an empty DataFrame to avoid further errors
    df = pd.DataFrame(columns=['timestamp', 'user_id', 'ip_address', 'country', 'success', 'os'])

# Display the first few rows of the DataFrame
print("\nDataFrame Head:")
display(df.head())

# Display information about the DataFrame, including data types and non-null values
print("\nDataFrame Info:")
df.info()
```

### 2. Data Cleaning

```python
# Check for missing values
print("\nMissing values per column BEFORE cleaning:")
print(df.isnull().sum())

# Convert 'timestamp' column to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])
print("\n'timestamp' column converted to datetime.")

# Create a new 'hour' column by extracting the hour from the timestamp
df['hour'] = df['timestamp'].dt.hour
print("\n'hour' column created.")

# Display the updated DataFrame info to confirm changes
print("\nDataFrame Info AFTER cleaning:")
df.info()

# Display the first few rows with the new 'hour' column
print("\nDataFrame Head AFTER cleaning:")
display(df.head())
```
